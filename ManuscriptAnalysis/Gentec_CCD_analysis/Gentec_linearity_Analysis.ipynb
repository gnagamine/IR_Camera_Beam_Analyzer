```python
import matplotlib.pyplot as plt
from SeriesAnalyzer_2 import SeriesAnalyzer_2
%matplotlib widget

measured_voltage_converted_from_lock_in_to_osciloscope_V_GENTEC = 1.19
angle_pol_voltage_measurement_GENTEC = 20
moved_pol = 'first'

GENTEC_horizontal_SeriesAnalyzer = SeriesAnalyzer_2(dir_path=dir_path_GENTEC_horizontal,
                                                    camera_name='gentec',
                                                    Y_ref_position_for_background_subtraction=2,
                                                    crop_range_x_pixels=crop_x_range_GENTEC_horizontal,
                                                    crop_range_y_pixels=crop_y_range_GENTEC_horizontal)

beam_characterization_df_GENTEC_horizontal = GENTEC_horizontal_SeriesAnalyzer.get_beam_char_df_w_powers(
    save_data_plot_bool=True,
    known_angle=angle_pol_voltage_measurement_GENTEC,
    kown_voltage_at_known_angle_in_V=measured_voltage_converted_from_lock_in_to_osciloscope_V_GENTEC,
    moved_polarizer=moved_pol)

beam_characterization_df_GENTEC_horizontal
```

```python
%matplotlib widget
from DataFitter import DataFitter

beam_characterization_df_GENTEC_horizontal_for_plot = beam_characterization_df_GENTEC_horizontal[
    beam_characterization_df_GENTEC_horizontal['Power (uW)'] <= 100]
fig_fitting_HIKMICRO, ax_fitting_HIKMICRO = DataFitter(model_function='quadratic_model').plot_data_with_fit(
    x_data=beam_characterization_df_GENTEC_horizontal_for_plot['Power (uW)'],
    y_data=beam_characterization_df_GENTEC_horizontal_for_plot['total_intensity'],
    x_label='Power (uW)',
    y_label='Intensity (a.u.)',
    title='Gentec horizontal',
)
fig_fitting_HIKMICRO.savefig('gentec_horizontal.pdf')
```
