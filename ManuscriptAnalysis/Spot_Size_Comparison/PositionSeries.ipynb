```python
import pandas as pd
import matplotlib.pyplot as plt
from SeriesAnalyzer import SeriesAnalyzer
%matplotlib inline

dir_path_HIKMICRO = '/Users/Shared/Files From c.localized/Gabriel_UniBern_Local/DataAnalysis/Low cost THz Camera/20250516/position series_HIKMICRO'
series_analyzer_HIKMICRO = SeriesAnalyzer(dir_path_HIKMICRO,
                                          bool_individual_backgrounds_taken=True,
                                          camera_name='HIKMICRO')
fitting_coefficients_df_HIKMICRO = series_analyzer_HIKMICRO.get_all_fitting_coefficients(bool_plot_2D_maps=False)
fitting_coefficients_df_HIKMICRO["position (mm)"] = (
    fitting_coefficients_df_HIKMICRO["filename"]
    .str.replace(",",
                 ".",
                 regex=False)
    .str.removesuffix(" mm.csv")
    .astype(float)
)
display(fitting_coefficients_df_HIKMICRO)

fig_HIKMICRO, ax_HIKMICRO = plt.subplots()
ax_HIKMICRO.plot(fitting_coefficients_df_HIKMICRO["position (mm)"],
                 fitting_coefficients_df_HIKMICRO["fwhm_x"],
                 'o',
                 label='fwhm_x',
                 )
ax_HIKMICRO.plot(fitting_coefficients_df_HIKMICRO["position (mm)"],
                 fitting_coefficients_df_HIKMICRO["fwhm_y"],
                 'o',
                 label='fwhm_y',
                 )
ax_HIKMICRO.set_xlabel("position (mm)")
ax_HIKMICRO.set_ylabel("fwhm (um)")
plt.show()
```
